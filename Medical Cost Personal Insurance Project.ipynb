```python
# We can see outliers present in BMI and Charges columns. We have remove the outliers using zscore and IQR techniques before building the model.
```

```python
from scipy.stats import zscore

out_features = df[['bmi','charges']]
z = np.abs(zscore(out_features))
z
```

```python
np.where(z>3)
```

```python
z.iloc[34,1]
```

```python
df1 = df[(z<3).all(axis=1)]
```

```python
df1.shape
```

```python
print("First DataFrame -", df.shape[0])
print("New DataFrame -", df1.shape[0])
```

```python
print("Data Loss percentage -",((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

```python
df=df1
```

```python
#The loss of data is 0 and it is affordable. So, now we can check with IQR method
```

```python
#1st Quantile
Q1 = out_features.quantile(0.25)
```

```python
#3RD Quantile
Q3 = out_features.quantile(0.75)
```

```python
#IQR
IQR = Q3 -Q1
df2 = df[~((df<(Q1 -1.5 *IQR))|(df>(Q3 + 1.5 * IQR))).any(axis =1)]
```

```python
df2.shape
```

```python
print (" Data Loss percentage after removing outliers with IQR method -", ((df.shape[0]-df2.shape[0])/df.shape[0])*100)
```

```python
#The data loss using IQR method is high compared to zscore method. So, we will go ahead with zscore value
```

```python
# Step -8, Normalization of data

from sklearn.preprocessing import StandardScaler
data_first = df1.copy()

Bmi = data_first.bmi
Bmi = Bmi.values.reshape(-1,1)
data_first['bmi'] = StandardScaler().fit_transform(Bmi)

Age = data_first.age
Age = Age.values.reshape(-1,1)
data_first['age'] = StandardScaler().fit_transform(Age)

Charges = data_first.charges
Charges = Charges.values.reshape(-1,1)
data_first['charges'] = StandardScaler().fit_transform(Charges)



```
